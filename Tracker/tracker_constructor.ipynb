```python
tape_noble = pd.read_csv('tapes/tape_noble-20200504.csv')
tape_patch = pd.read_csv('tapes/tape_patch-20200504.csv')
tape_ppm = pd.read_csv('tapes/tape_ppm-20200504.csv')
```

```python
def clean_tape(tape):
    tape_filename = 'tapes/tape_ppm-20200504.csv'
    
    tape.columns = tape.columns.str.strip()
    tape.columns = tape.columns.str.replace(' ', '_')
    tape.columns = tape.columns.str.lower()
    
    #lender
    lender = re.search('tapes/tape_(.*)-', tape_filename)
    lender = lender.group(1)
    
    tape['lender'] = lender
    
    #loan_id
    cols = ['loan_number', 'loan_id', 'loan_#']
    for col in cols:
        tape = tape.rename(columns={col:'loan_id'})
    
    #prop_address
    
    #prop_city
    
    #prop_state
    
    #prop_zip
    
    #prop_aiv
    cols = ['estimated_as-is_value', 'as-is_value', 'total_cost', 'market_value']
    for col in cols:
        tape = tape.rename(columns={col:'prop_aiv'})
    
    #prop_arv
    cols = ['']
    for col in cols:
        tape = tape.rename(columns={col:'prop_arv'})
        
    #prop_type
    cols = ['property_type']
    for col in cols:
        tape = tape.rename(columns={col:'prop_type'})
    
    #prop_cost_basis
    cols = ['total_purchase_price', 'total_cost']
    for col in cols:
        tape = tape.rename(columns={col:'prop_cost_basis'})
    
    #bor_name
    cols = ['primary_guarantor_name', 'borrowing_entity']
    for col in cols:
        tape = tape.rename(columns={col:'bor_name'})
    
    #bor_address
    cols = ['']
    for col in cols:
        tape = tape.rename(columns={col:'bor_address'})
    
    #bor_city
    cols = ['']
    for col in cols:
        tape = tape.rename(columns={col:'bor_city'})
    
    #bor_state
    cols = ['']
    for col in cols:
        tape = tape.rename(columns={col:'bor_state'})
        
    #bor_zip
    cols = ['']
    for col in cols:
        tape = tape.rename(columns={col:'bor_zip'})
    
    #bor_fico
    cols = ['borrower_fico', 'credit']
    for col in cols:
        tape = tape.rename(columns={col:'bor_fico'})
    
    #bor_liq
    cols = ['']
    for col in cols:
        tape = tape.rename(columns={col:'bor_liq'})
    
    #loan_orig_date
    cols = ['origination_date', 'close']
    for col in cols:
        tape = tape.rename(columns={col:'loan_orig_date'})    
    
    #loan_mat_date
    cols = ['maturity']
    for col in cols:
        tape = tape.rename(columns={col:'loan_mat_date'})    
    
    #loan_term
    cols = ['loan_term', 'term']
    for col in cols:
        tape = tape.rename(columns={col:'loan_term'})    
    
    #loan_opb
    cols = ['loan_amount', 'original_loan_amount']
    for col in cols:
        tape = tape.rename(columns={col:'loan_opb'})
    
    #loan_upb
    cols = ['current_balance']
    for col in cols:
        tape = tape.rename(columns={col:'loan_upb'})    
    
    #loan_position
    cols = ['lien_status']
    for col in cols:
        tape = tape.rename(columns={col:'loan_position'})
    
    #loan_ltc_ratio_stated
    cols = ['ltc']
    for col in cols:
        tape = tape.rename(columns={col:'loan_ltc_ratio_stated'})
    
    #loan_ltaiv_ratio_stated
    cols = ['']
    for col in cols:
        tape = tape.rename(columns={col:'loan_ltaiv_ratio_stated'})
    
    #loan_ltarv_ratio_stated
    cols = ['']
    for col in cols:
        tape = tape.rename(columns={col:'loan_ltarv_ratio_stated'})
    
    #loan_exit_strat
    cols = ['']
    for col in cols:
        tape = tape.rename(columns={col:'loan_exit_strat'})
    
    #loan_type
    cols = ['']
    for col in cols:
        tape = tape.rename(columns={col:'loan_type'})
    
    #draw_requested
    cols = ['']
    for col in cols:
        tape = tape.rename(columns={col:'draw_requested'})
    
    #draw_total
    cols = ['']
    for col in cols:
        tape = tape.rename(columns={col:'draw_total'})
    
    #draw_remaining
    cols = ['']
    for col in cols:
        tape = tape.rename(columns={col:'draw_remaining'})
    
    #draw_disbursed
    cols = ['']
    for col in cols:
        tape = tape.rename(columns={col:'draw_disbursed'})
    
    return tape

```

```python
tapes = [tape_noble, tape_patch, tape_ppm]
```

```python
for i in range(len(tapes)):
    tapes[i] = clean_tape(tapes[i])
```
